# Data Analysis and Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
x = ['Not Cancelled','Cancelled'] # to make is easier to understand
plt.figure(figsize = (5,4))
plt.title('Reservation satus cont')
plt.bar(x,df['is_canceled'].value_counts())
plt.show()
```

```python

plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
plt.title('Reservation status is different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')['adr'].mean()
city_hotel = city_hotel.groupby('reservation_status_date')['adr'].mean()
```

```python
resort_hotel
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel, label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel, label = 'City Hotel')
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
plt.title('Reservation status per month', fontsize = 30)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')['adr'].mean()
print(cancelled_df_adr)
cancelled_df_adr.reset_index()
cancelled_df_adr.sort_values()
```

```python
not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')['adr'].mean()
print('not_cancelled_data')
not_cancelled_df_adr.reset_index()
not_cancelled_df_adr.sort_values()
```

```python
plt.figure(figsize = (20,5))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(not_cancelled_df_adr.index, not_cancelled_df_adr, label = 'not cancelled')
plt.plot(cancelled_df_adr.index, cancelled_df_adr, label = 'cancelled')
plt.legend(fontsize = 20)
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr.index > '2016') & (cancelled_df_adr.index < '2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr.index > '2016') & (not_cancelled_df_adr.index < '2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title("Average Daily Rate", fontsize = 30)

plt.plot(not_cancelled_df_adr.index, not_cancelled_df_adr, label = 'not cancelled')
plt.plot(cancelled_df_adr.index, cancelled_df_adr, label = 'cancelled')

plt.legend(fontsize = 20)
plt.show()
```
